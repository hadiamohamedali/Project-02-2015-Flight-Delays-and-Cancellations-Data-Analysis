```python
# Visualization For IATA Code
plt.figure(figsize = (12 , 6))
sns.countplot(y = 'Iata Code' , data = data1 , palette = 'viridis')
plt.title('Distribution of IATA Codes.')
plt.xlabel('Count')
plt.ylabel('IATA Code')
plt.legend()
plt.show()
```

```python
# Visualization For Airlines
plt.figure(figsize = (12 , 6))
sns.countplot(y = 'Airline' , data = data1 , palette = 'magma')
plt.title('Distribution of Airlines.')
plt.xlabel('Count')
plt.ylabel('Airline')
plt.legend()
plt.show()
```

```python
mapping = pd.crosstab(data1['Iata Code'] , data1['Airline'])

plt.figure(figsize = (12 , 8))
sns.heatmap(mapping , cmap = 'viridis' , cbar = False)

plt.title('Relationship Between IATA Code in Airlines')
plt.xlabel('Airline')
plt.ylabel('IATA Code')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.axis('off')

table_data = data1[['Iata Code', 'Airline']]
plt.table(
    cellText=table_data.values,
    colLabels=table_data.columns,
    loc='center'
)

plt.title('IATA Code to Airline Mapping')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(
    y='Airline',
    data=data1,
    palette='viridis'
)

plt.title('Airlines and their IATA Codes')
plt.xlabel('Count')
plt.ylabel('Airline')
plt.show()
```


The relationship between IATA Code and Airline is a one-to-one mapping, therefore visualizations are used for clarification and representation rather than statistical correlation.

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Airline', data=data3, palette='viridis')
plt.title('Number of Flights per Airline')
plt.xlabel('Airline (IATA Code)')
plt.ylabel('Number of Flights')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Day_of_week', data=data3, palette='coolwarm')
plt.title('Flights per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Flights')
plt.show()
```

```python
top_origins = data3['Origin_airport'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_origins.values, y=top_origins.index, palette='magma')
plt.title('Top 10 Origin Airports')
plt.xlabel('Number of Flights')
plt.ylabel('Airport')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Cancelled', data=data3, palette='Set2')
plt.title('Cancelled vs Completed Flights')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data3['Departure_delay'].dropna(), bins=100, kde=True, color='skyblue')
plt.xlim(-50, 500)
plt.title('Departure Delay Distribution')
plt.xlabel('Minutes')
plt.ylabel('Number of Flights')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data3['Arrival_delay'].dropna(), bins=100, kde=True, color='salmon')
plt.xlim(-50, 500)
plt.title('Arrival Delay Distribution')
plt.xlabel('Minutes')
plt.ylabel('Number of Flights')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data3['Distance'].dropna(), bins=50, kde=True, color='lightgreen')
plt.title('Flight Distance Distribution')
plt.xlabel('Miles')
plt.ylabel('Number of Flights')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Distance', y='Arrival_delay', data=data3, alpha=0.3)
plt.title('Distance vs Arrival Delay')
plt.xlabel('Distance (Miles)')
plt.ylabel('Arrival Delay (Minutes)')
plt.show()
```

```python
cancel_reason_counts = data3['Cancellation_reason'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=cancel_reason_counts.index, y=cancel_reason_counts.values, palette='magma')
plt.title('Cancellation Reasons')
plt.xlabel('Reason Code')
plt.ylabel('Number of Flights')
plt.show()
```

```python
delay_cols = ['Air_system_delay','Security_delay','Airline_delay','Late_aircraft_delay','Weather_delay']

plt.figure(figsize=(10,6))
data3[delay_cols].sum().plot(kind='bar', color='skyblue')
plt.title('Total Delay Minutes by Cause')
plt.ylabel('Minutes')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Month', data=data3, palette='coolwarm')
plt.title('Number of Flights per Month')
plt.xlabel('Month')
plt.ylabel('Number of Flights')
plt.show()
```

```python
avg_monthly_delay = data3.groupby('Month')['Arrival_delay'].mean()
plt.figure(figsize=(10,5))
sns.lineplot(x=avg_monthly_delay.index, y=avg_monthly_delay.values, marker='o', color='salmon')
plt.title('Average Arrival Delay per Month')
plt.xlabel('Month')
plt.ylabel('Average Delay (Minutes)')
plt.show()
```


Multiple visualization techniques were applied to each column across Airlines, Airports, and Flights datasets to highlight distribution, frequency, density, and geographical patterns. These visualizations provide a clear and comprehensive understanding of the 2015 Flight Delays and Cancellations data.
